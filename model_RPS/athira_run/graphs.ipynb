```julia
using CSV
using DataFrames
using Plots
```

```julia
transmission = CSV.read("./results_western_4weeks/transmission_results.csv", DataFrame)
```

```julia
lines = transmission.Line
total_transfer_capacity = transmission.Total_Transfer_Capacity
start_transfer_capacity = transmission.Start_Transfer_Capacity
change_in_transfer_capacity = transmission.Change_in_Transfer_Capacity
```

```julia
# Bar plot for start and total transfer capacities
bar(lines, [start_transfer_capacity total_transfer_capacity],
    label=["Start Capacity" "Total Capacity"], 
    bar_width=0.4, color=[:green :blue], alpha=0.8)

# Overlay line plot for the change in transfer capacity
plot!(lines, change_in_transfer_capacity, 
      seriestype=:scatter, 
      label="Change in Capacity", 
      color=:red, 
      markershape=:circle, linewidth=2)

# Add labels and title
xlabel!("Line")
ylabel!("Transfer Capacity")
title!("Transmission Results")
```
